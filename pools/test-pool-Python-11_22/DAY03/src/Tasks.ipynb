# Задание 1
Напишите скрипт, решающий задачу, описанную по адресу https://contest.yandex.ru/contest/40435/problems/14/.

```python
class Calculator:
    result = 0
    last = [None, None]

    def sum(self, a, b):
        result = a + b
        self.last[1] = self.last[0]
        self.last[0] = self.message("sum", a, b, result)
        return result
        
    def sub(self, a, b):
        result = a - b
        self.last[1] = self.last[0]
        self.last[0] = self.message("sub", a, b, result)
        return result

    def mul(self, a, b):
        result = a * b
        self.last[1] = self.last[0]
        self.last[0] = self.message("mul", a, b, result)
        return result

    def div(self, a, b, mod=False):
        self.last[1] = self.last[0]
        if (mod):
            result = a % b
        else:
            result = a / b
        self.last[0] = self.message("div", a, b, result, mod)
        return result

    def history(self, n):
        return self.last[n - 1]

    def message(self, oper, a, b, res, mod=False):
        if(mod):
            string = f'{oper}({a}, {b}, {mod}) # == {res}'
        else:
            string = f'{oper}({a}, {b}) # # == {res}'
        return string
    
    @classmethod
    def clear(cls):
        cls.last = [None, None]
```

```python
cal = Calculator()
print( 1, 1, 1, 1,cal.sum(5,5))
print(cal.sub(5,5))
print(cal.history(1))
print(cal.history(2))
cal.clear()
print(cal.last)
print(cal.mul(5,5))
print(cal.div(5,5))
print(cal.div(5,5, True))
print(cal.history(1))
print(cal.history(2))
```

# Задание 2
Напишите скрипт, решающий задачу, описанную по адресу https://contest.yandex.ru/contest/40435/problems/15/.

```python
class BaseWallet:
    def __init__(self, name, amounth):
        self.name = name
        self.amounth = amounth
    
    def __add__(self, other):
        if (isinstance(self, BaseWallet) and isinstance(other, BaseWallet)):
            coef = (other.exchange_rate[1] / self.exchange_rate[1])
            self.amounth = int(self.amounth + (other.amounth * coef))
        else: 
            self.amounth = self.amounth + other
        return self
    
    def __sub__(self, other):
        if (isinstance(self, BaseWallet) and isinstance(other, BaseWallet)):
            coef = (other.exchange_rate[1] / self.exchange_rate[1])
            self.amounth = int(self.amounth - (other.amounth * coef))
        else: 
            self.amounth = self.amounth - other
        return self
    
    def __mul__(self, other):
        if (isinstance(self, BaseWallet) and isinstance(other, BaseWallet)):
            coef = (other.exchange_rate[1] / self.exchange_rate[1])
            self.amounth = int(self.amounth * (other.amounth * coef))
        else: 
            self.amounth = self.amounth * other
        return self

    def __iadd__(self, other):
        self + other
        return self
        
    def __isub__(self, other):
        self - other
        return self
    
    def __imul__(self, other):
        self * other
        return self
    
    def __radd__(self, other):
        self + other
        return self
    
    def __rsub__(self, other):
        self - other
        return self
    
    def __rmul__(self, other):
        self * other
        return self
    
    def __eq__(self, other):
        return self.exchange_rate[0] == other.exchange_rate[0] and self.amounth == other.amounth

    def spend_all(self):
        if self.amounth > 0: self.amounth = 0
    
    def to_base(self):
        return self.amounth * self.exchange_rate[1]

    def __str__(self):
        return f'{self.exchange_rate[0]} Wallet {self.name} {self.amounth}'

class RubbleWallet(BaseWallet):
    exchange_rate = ('Rubble', 1, )
    def __init__(self, name, amounth):
        super(RubbleWallet, self).__init__(name, amounth)



class DollarWallet(BaseWallet):
    exchange_rate = ('Dollar', 60, )
    def __init__(self, name, amounth):
        super(DollarWallet, self).__init__(name, amounth)



class EuroWallet(BaseWallet):
    exchange_rate = ('Euro', 70, )
    def __init__(self, name, amounth):
        super(EuroWallet, self).__init__(name, amounth)
        
```
